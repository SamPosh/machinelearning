```python
titanic_data.replace({"gender":{"male":1,"female":2}},inplace=True)
```

```python
titanic_data
```

```python
input_data = titanic_data.drop(columns=["name","cabin","ticket","embarked","survived"])
out_data = titanic_data["survived"]
print(input_data)
print(out_data)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
input_data = scaler.fit_transform(input_data)
print(input_data)
```

```python
print(input_data.shape)
print(out_data.shape)
```

```python
from sklearn.model_selection import train_test_split
input_data_train,input_data_test,out_data_train,out_data_test= train_test_split(input_data,out_data,test_size=0.3)
```

```python
print(input_data_train.shape)
print(input_data_test.shape)
print(out_data_train.shape)
print(out_data_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(input_data_train,out_data_train)
predicted = model.predict(input_data_test)
predicted
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(out_data_test,predicted)
```

```python
import pandas as pd
df = pd.DataFrame(confusion_matrix(out_data_test,predicted),columns=['predicted_no','predicted_yes'], index=['actual_no','actual_yes'])
df
```

```python
total_predicted_no = df["predicted_no"].sum()
print(f"Actually predicted as No: {total_predicted_no}")
# Actual No is number of people actually died

total_predicted_yes = df["predicted_yes"].sum()
print(f"Actually predicted yes : {total_predicted_yes}")
# Actual yes is number of people actually survived .
```

```python
from sklearn.metrics import accuracy_score
accuracy_info = accuracy_score(out_data_test,predicted)
accuracy_info
```
